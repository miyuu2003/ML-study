### NumPy 配列の基本
配列の形（shape）と型（dtype）を確認  
shape：配列の“サイズ情報”  
dtype：要素の型


```python
x = np.array([1.0, 2.0, 3.0])
print("x:", x)
print("shape:", x.shape)
print("dtype:", x.dtype)
```

### 行列（2次元配列）

shape = (行数, 列数) 

```python
A = np.array([[1, 2], [3, 4]])
print(A)
print("shape:", A.shape)
```

### 要素ごとの演算

```python
B = np.array([[10, 20], [30, 40]])
print("A + B:\n", A + B)
print("A * B:\n", A * B)
```

### 行列積

@ は 行列積（行列の掛け算）  
C は単位行列なので、A @ C = A

```python
C = np.array([[1, 0], [0, 1]])
print("A @ C:\n", A @ C)
```

### ブロードキャスト

A は (2,2)、v は (2,)であるが、ブロードキャストを用いて拡張  
バイアス加算で使われることが多い

```python
v = np.array([10, 20])
print("A + v:\n", A + v)
```

## まとめ
- NumPy配列は数式のかたまり
- shape を意識するのが大事
- 行列積（@）がNNの基本計算
- ブロードキャストは暗黙的に形を合わせてくれる
